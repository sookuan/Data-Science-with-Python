## Working with lists

```python
# below is a list of 10 students (using just first name)
students = ['Alice','Ben','Cathy','Dawn','Emily','Fern','Emma','Sean','Felix','Hugh']
```

```python
# Print the list
print(f"Students: {students}")
```

```python
# Print the list of students in the following format
# Index 01 	 Alice
# Index 02 	 Ben
# Index 03 	 Cathy
# ...
for index, name in enumerate(students): #enumerate: itereate over a sequence
    print(f"Index 0{index+1} {name}")
```

```python
# print the list of students in reverse order one student in a line with its index
# google search how to print list in reverse order
for i in reversed(range(len(students))):
    print(f"Index 0{i+1} {students[i]}")
```

```python
#Alternative
students_reversed = students.copy() #copy the list of students and save as new variable
students_reversed.reverse()
for i, student in enumerate(students_reversed):
    print(f'Index {i+1:02d} \t {student}') # 02d = 01,02... , \t = tab
```

```python
# print the student list in random order
# hint look up the function random.sample
import random
random_student = random.sample(students , len(students)) # with the size of sample = len(students)
for i , student in enumerate(random_student):
    print(f"Index {i+1:02d} \t {student}")
```

```python
# list students in decreasing order of their exam marks given their marks below
# hint: look up the 'sorted' function
marks = [87,45,76,64,66,90,55,73,69,75]
studentsAndMarks = dict(zip(students,marks)) # convert to dictionary with key:value
print(f"Students and their marks: \n {studentsAndMarks}")

def value_getter(item):
    return item[0]
print(f"student and their marks :\n {studentsAndMarks}")
for student in sorted(studentsAndMarks, key =studentsAndMarks.get, reverse=True): # studentAndMarks.get = use mark to sort
    print(f"{student:10s}:{studentsAndMarks[student]}")
```

```python
#desired output:
sort_list = sorted(studentsAndMarks.items(), key=lambda x: x[1], reverse=True)

for index, (student, mark) in enumerate(sort_list, start=1):
    print(f"Rank {index:02d} \t {student}: {mark}")
```

## Using matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a set of x values
x = np.linspace(0, 10, 101) # range 1 - 10 with 101 values
# Print the inputs
print(x)
```

```python
# Create a figure using plt.figure
#fig = plt.figure()
# Plot the sine of x
plt.plot(x, np.sin(x), '-')
# Plot the cosine of x
plt.plot(x, np.cos(x), '.')
plt.show()
```

## Plotting bar charts

```python
# Create two lists of 5 arbitrary numbers
array1 = [2, 7, 1, 9, 4]
array2 = [5, 3, 8, 6, 0]

# Plot two bar charts using plt.bar and the lists you've created above
# Label them "Example one" and "example two"

plt.bar(np.arange(2*len(array1))[::2]+1, array1, label="Example one", width=0.4)
plt.bar(np.arange(2*len(array1))[::2], array2, label="Example two", color='g', width=0.4)

plt.legend()
# Label x axis "Bar number"
plt.xlabel('X')
# Label y axis "Bar height"
plt.ylabel('Y')
# Set title "Bar Graph"
plt.title("Bar Graph")
# Show the chart
plt.show()
```

```python
# Histogram Code
# Create an array of a random Laplace distribution (a type of random number generator)
test = np.random.laplace(loc=15, scale=3, size=100)
# Create a histogram using plt.hist()
n, bins, patches = plt.hist(x=test, bins='auto', color='#0504aa',
                            alpha=0.7, rwidth=0.85)
# Set grid for x and y
# set transperancy of the grid using alpha=0.5
plt.grid(axis='both', alpha = 0.5)
plt.grid(axis='x', alpha = 0.5)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('My First Histogram Ever')
plt.text(23, 45, r'$\mu=15, b=3$')
maxfreq = n.max()
# Set a clean upper y-axis limit.
plt.ylim(ymax=np.ceil(maxfreq / 10) * 10 if maxfreq % 10 else maxfreq + 10)
plt.show()
```

```python
# Bonus: Create a scatter plot of random data
np.random.seed(42) #repeatability
y = np.random.rand(50)
x = np.random.rand(50)

plt.figure(figsize = (8,6))
plt.scatter(x, y, alpha=0.7, c='green', label='Random Plot')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Scatter Plot of 50 Random Data')
plt.legend()
plt.show()
```

## Visualising the iris dataset with seaborn

```python
# Import seaborn library
import seaborn as sns
# Import load_iris
from sklearn.datasets import load_iris
# Load the dataset
dataset=load_iris()
# Place in pandas dataframe
data=pd.DataFrame(dataset["data"],columns=["Petal length","Petal Width","Sepal Length","Sepal Width"])
data["Species"]=dataset["target"]
data["Species"]=data["Species"].apply(lambda x: dataset["target_names"][x]) # map species name to 0,1,2...
# Print top 10 rows
print(data.head(10))
```

```python
# Describe the dataframe
print(data.describe())
```

```python
# use seaborn pairplot function to plot relationship between variables
sns.pairplot(data, hue="Species")
```
